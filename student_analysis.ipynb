# Loading Dataset

```python
import pandas as pd
df= pd.read_csv("student-mat.csv", sep=";")
df.head()
```

## Exploring & Cleaning Data

##### Dataset Information

```python
print("Shape: ",df.shape)
```

```python
df.info()
```

```python
df.describe()
```

##### Missing Values

```python
df.isnull()
df.isnull().sum()
```

##### Remove Duplicates

```python
print("Duplicates:", df.duplicated().sum())

df = df.drop_duplicates()
```

## Analysis Questions

##### Average final grade (G3)

```python
print("Average final grade (G3) is : " , df["G3"].mean())
```

##### Number of Students Scoring Above 15

```python
print("Number of Students Scoring Above 15 are: ", (df["G3"] >15).sum())
```

##### Correlation between Study Time and Performance

```python
df["studytime"].corr(df["G3"])
```

##### Gender Performance

```python
df.groupby("sex")["G3"].mean()
```

## Visualizations

##### Histogram of grades

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['G3'], bins=10 , kde= True , color="royalblue")
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade(G3)")
plt.ylabel("Count")
plt.show()
```

##### Scatterplot: Study Time vs Grades

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x= "studytime", y="G3", data = df,color= 'maroon')
plt.title("Study Time Vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

##### Male vs Female Average Score

```python
avg_scores = df.groupby("sex")["G3"].mean().reset_index()
print(avg_scores)
plt.figure(figsize=(8,5))

sns.barplot(data= avg_scores, x= "sex" , y= "G3", color = "seagreen")
plt.title("Average Final Grade by Gender ")
plt.xlabel("Gender")
plt.ylabel("Average Grade(G3)")
plt.show()
```
